```python
import hashlib

class MerkleTree:

    def myhash(self, data):
        return hashlib.sha256(data.encode()).hexdigest()

    def my_merkle_tree(self, data_blocks):
        if len(data_blocks) % 2 != 0:
            data_blocks.append(data_blocks[-1])

        # Hash each data block
        hashed_blocks = [self.myhash(data) for data in data_blocks]

        # Construct the Merkle Tree
        while len(hashed_blocks) > 1:
            new_hashes = [self.myhash(a + b) for a, b in zip(hashed_blocks[::2], hashed_blocks[1::2])]
            
            # If there's an odd number of blocks, append the last block
            if len(hashed_blocks) % 2 != 0:
                new_hashes.append(hashed_blocks[-1])
            
            hashed_blocks = new_hashes

        return hashed_blocks[0]

# Example usage:
file_path = "./client_storage/new.txt"
with open(file_path, 'r') as file:
    my_file = file.read().splitlines()

print(my_file)

obj = MerkleTree()
out = obj.my_merkle_tree(my_file)
print(out)
```

```python
import os

def encrypt(data, key):
    encrypted_data = bytearray()
    for i in range(len(data)):
        char = data[i]
        key_char = key[i % len(key)]
        encrypted_char = (char + key_char) % 256
        encrypted_data.append(encrypted_char)
    return bytes(encrypted_data)

def decrypt(data, key):
    decrypted_data = bytearray()
    for i in range(len(data)):
        char = data[i]
        key_char = key[i % len(key)]
        decrypted_char = (char - key_char) % 256
        decrypted_data.append(decrypted_char)
    return bytes(decrypted_data).decode()

def generate_key(length):
    return os.urandom(length)

def read_file(filename):
    with open(filename, 'rb') as file:
        return file.read().decode()

def write_file(filename, content):
    with open(filename, 'wb') as file:
        file.write(content.encode())

# Example: Generate a 16-byte (128-bit) key
secure_key = generate_key(16)
print(f"Secure Key: {secure_key}")

# Example: Read content from a file
filename = './client_storage/new.txt'
original_content = read_file(filename)
print(f"Original Content: {original_content}")

# Example: Encrypt the content
encrypted_content = encrypt(original_content.encode(), secure_key)
print(f"Encrypted Content: {encrypted_content}")

# Example: Decrypt the content
decrypted_content = decrypt(encrypted_content, secure_key)
print(f"Decrypted Content: {decrypted_content}")

# Example: Verify if encryption and decryption are successful
if original_content.strip() == decrypted_content.strip():
    print("True")
    print(original_content)
```
